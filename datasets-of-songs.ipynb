```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
```

```python
df_genre=pd.read_csv('archive/genres_v2.csv')
df_genre.head()
```

```python
genre_columns=df_genre.columns
print(genre_columns)
```

```python
df_genre.describe(include='all')
```

```python
songsId=df_genre['id']
songsId
```

```python
df_genre.drop(['type'],axis=1,inplace=True)
```

```python
df_genre.drop(['id'],axis=1,inplace=True)
```

```python
df_genre.drop(['uri', 'track_href', 'analysis_url', 'duration_ms'],axis=1,inplace=True)
```

```python
df_genre.drop([ 'song_name', 'Unnamed: 0', 'title'],axis=1,inplace=True)
```

```python
df_genre.head()
```

```python
df_genre.genre.value_counts()
```

```python
df_genre.genre.unique()
```

```python
genre_list=list(df_genre.genre.unique())
genre_list
```

```python
genres={}
for i in range(len(genre_list)):
    genres[genre_list[i]]=i
    
print(genres)
```

```python
genreList=[]
for i in df_genre.genre:
    if i in genre_list:
        genreList.append(genres[i])
```

```python
df_genre['Genres']=genreList
```

```python
df_genre.drop(['genre'],inplace=True,axis=1)
```

```python
print(df_genre.columns)
```

```python
Y=df_genre.Genres.values
```

```python
df_genre.dtypes
```

```python
df_genre['mode']=df_genre['mode'].astype('str')
```

```python
df_genre['key']=df_genre['key'].astype('str')
```

```python
df_genre['time_signature']=df_genre['time_signature'].astype('str')
```

```python
df_genre.columns
```

```python
df_genre2=df_genre
df_genre2.head()
```

```python
df_genre2.drop(['Genres','time_signature'],inplace=True,axis=1)
```

```python
df_genre2.head()
```

```python
X=df_genre2.values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print('X_train Shape: {}'.format(X_train.shape))
print('X_test Shape: {}'.format(X_test.shape))
print('Y_train Shape: {}'.format(y_train.shape))
print('Y_test Shape: {}'.format(y_test.shape))
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=1000)
clf.fit(X_train_scaled,y_train)
print(clf.score(X_test_scaled,y_test))
```

```python
y_pred=clf.predict(X_test_scaled)
y_pred
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
fig,ax=plt.subplots(figsize=(20,10))
ax=sns.heatmap(cm,annot=True,fmt='')

ax.set_xlabel('Orginal Label')
ax.set_ylabel('Predicted Label')
```
